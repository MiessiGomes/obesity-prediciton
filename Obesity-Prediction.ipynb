```python
import scipy.stats as stats
import numpy as np

cols = ['Gender', 'FAVC', 'CAEC', 'SMOKE', 'SCC', 'CALC', 'MTRANS']

for col in cols:
    crosstab = pd.crosstab(cat[col], cat['HistoryOverweight'])


    chi2, p, dof, expected = stats.chi2_contingency(crosstab)

    # Coeficiente de Cramér
    n = crosstab.sum().sum()
    phi2 = chi2/n
    r,k = crosstab.shape
    phi2corr = max(0, phi2 - ((k-1)*(r-1))/(n-1))
    rcorr = r - ((r-1)**2)/(n-1)
    kcorr = k - ((k-1)**2)/(n-1)
    cramers_v = np.sqrt(phi2corr / min((kcorr-1), (rcorr-1)))

    print(f"Coluna: {col}")
    print(f"Valor de p: {p}")
    print(f"Coeficiente de Cramér: {cramers_v}")
    print("\n")
```
